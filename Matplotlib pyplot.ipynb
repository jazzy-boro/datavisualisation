```python
import matplotlib.pyplot as plt
import numpy as np
xpoint=np.array([1,2])
ypoint=np.array([4,3])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
plt.plot(xpoint,ypoint,'o')
plt.show()
```

```python
xpoint=np.array([1,2,9,7,9])
ypoint=np.array([4,3,8,4,7])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
plt.plot(xpoint,ypoint,marker='o')
plt.show()
```

```python
plt.plot(xpoint,ypoint,marker='*')
plt.show()
```

```python
plt.plot(xpoint,ypoint,linestyle='dotted')
plt.show()
```

```python
plt.plot(xpoint,ypoint,marker='o',color='pink')
plt.show()
```

```python
y1=np.array([1,2,3,4,6,6,3])
y2=np.array([7,3,5,7,3,9,8])
plt.plot(y1,color='red')
plt.plot(y2,color='green')
plt.show()
```

```python
x=np.array([21,12,345,667,24,787,909])
y=np.array([5,4,20,40,6,50,80])
plt.plot(x,y,linestyle='dashed',color='black')
plt.title('exercise')
plt.xlabel("calories burn")
plt.ylabel("duration")
plt.show()
```
